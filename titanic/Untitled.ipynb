```python
print(df[['Cabin', 'Age']])
```

```python
data=df.drop(df.columns[-2], axis=1)
data
```

```python
data.describe()
```

```python
data.Age = data.Age.fillna(data.Age.mean())
data
```

```python
# changing the age column from float to int
data['Age']=data['Age'].astype(int)
#changing the Fare column float to 2 decimal to place it in monertery terms
data['Fare']=data['Fare'].astype(float).round(2)
```

```python
data.info()
```

```python
# Checking duplicates values 
print(data.duplicated(keep=False))
data
```

```python
sns.countplot(data['Survived'])
plt.xticks([0, 1], ['Died', 'Alive'])
plt.show()
```

```python
ax=data['Age'].value_counts()\
                .head(15)\
                .plot(kind='bar', title='Age  Bar Presentation')
ax.set_ylabel("count")
ax.set_xlabel("Age")
```

```python
ax=data['Pclass'].value_counts()\
                .plot(kind='bar', title='Passenger class  Bar Presentation')
ax.set_ylabel("count")
ax.set_xlabel("Pclass")
```

```python
ax=data['Fare'].value_counts()\
                .head(5)\
                .plot(kind='bar', title='FARE Bar Presentation')
ax.set_ylabel("count")
ax.set_xlabel("Fare")
```

```python
ax=data['Sex'].value_counts()\
                .head(15)\
                .plot(kind='bar', title='Sex  Bar Presentation')
ax.set_ylabel("count")
ax.set_xlabel("Sex")
```

```python
def plot_hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(df[variable], bins=50)
    plt.xlabel(variable)
    plt.ylabel('count')
    plt.title('{} distribution with hist'.format(variable))
    plt.show          



numeric=['Fare', 'Age']
for  n in numeric:
    plot_hist(n)
```

```python
cate_value=[]
cont_value=[]

for column in data.columns:
    if data[column].nunique() <=10:
        cate_value.append(column)
    else:
        cont_value.append(column)
```
